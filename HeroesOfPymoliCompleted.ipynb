```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
PlayerCount = purchase_data['SN'].count()
print(f"The Player Count is {PlayerCount}!")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data['Item Name'].nunique()
avgpurchase = float(purchase_data['Price'].mean())
rev = purchase_data['Price'].sum()
purchasenum = purchase_data['Price'].count()
pd.options.display.float_format = '${:,.2f}'.format
main = {'Unique Items': [unique_items],
        'Number of Purchases': [purchasenum],
       'Average Purchase': [avgpurchase],
       'Total Revenue': [rev]}
maindf = pd.DataFrame(data=main)
maindf
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
pd.options.display.float_format = '{:,.0f}%'.format
genderdf = purchase_data.groupby('Gender').count()
genderdf = genderdf.drop(columns = "SN")
genderdf = genderdf.drop(columns = "Age")
genderdf = genderdf.drop(columns = "Item ID")
genderdf = genderdf.drop(columns = "Price")
genderdf = genderdf.drop(columns = "Item Name")
Totalgender = genderdf['Purchase ID'].sum()
Femalecount = genderdf.loc['Female', 'Purchase ID']
Malecount = genderdf.loc['Male', 'Purchase ID']
Othercount = genderdf.loc['Other / Non-Disclosed', 'Purchase ID']
Femaleperc = round(Femalecount / Totalgender * 100)
Maleperc = round(Malecount / Totalgender * 100)
Otherperc = round(Othercount / Totalgender * 100)
genderdemo = pd.DataFrame([Femalecount, Malecount, Othercount],
                          index=['Female', 'Male', 'Other/Non-Disclosed'],
                          columns=['Count of Genders'])
genderdemo1 = genderdemo
genderdemo1['Percent'] = [Femaleperc, Maleperc, Otherperc]

genderdemo1
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pd.options.display.float_format = '{:,.2f}'.format
genderanalysis = genderdemo.drop(columns = 'Percent')
genderanalysis.rename(columns = {'Count of Genders': "Number of Purchases"})
avgdf = purchase_data.groupby('Gender').mean()
avgdf = avgdf.drop(columns = "Purchase ID")
avgdf = avgdf.drop(columns = "Item ID")
totalpurchasevalue = purchase_data.groupby("Gender").sum()
totalpurchasevalue = totalpurchasevalue.drop(columns = ['Item ID', 'Age', 'Purchase ID'])
avgdf['Total Purchase Value'] = totalpurchasevalue['Price']
avgdf
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
pd.options.display.float_format = '{:,.2f}%'.format
agebin = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 99999]
agelabel = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data['Age Demographic'] =  pd.cut(purchase_data['Age'], bins = agebin, labels = agelabel, include_lowest=True)
agegroupby = purchase_data.groupby("Age Demographic")
agecount = agegroupby['SN'].nunique()
ageper = (agecount / PlayerCount) * 100
agedf = pd.DataFrame({"Count of Demographic": agecount,
                     "Percent of Demographic": ageper})
agedf
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
groupbyspent = purchase_data.groupby('SN')
numofspending = groupbyspent['Purchase ID'].count()
totalspending = groupbyspent['Price'].sum()
averagespending = groupbyspent['Price'].mean()

big_spenders = pd.DataFrame({"Purchase Count": numofspending,
                           "Total Spent": totalspending,
                           "Average Ammount Spent": averagespending})
big_spenderscomplete = big_spenders.sort_values(['Total Spent'], ascending=False).head(5)
big_spenderscomplete
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = purchase_data[['Item ID', 'Price', 'Item Name']]

iteminfo = items.groupby(['Item ID', 'Item Name'])

purchasesum = iteminfo['Price'].sum()

purchasenum = iteminfo['Price'].count()

itemprice = purchasesum / purchasenum

popitems = pd.DataFrame({"Item Price": itemprice,
                         "Times Purchased": purchasenum,
                         "Revenue From Item": purchasesum
                        })
popitemscomplete = popitems.sort_values(["Times Purchased"], ascending=False,)
popitemscomplete
```
